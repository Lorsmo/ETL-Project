```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from config import api_key
```

```python
# Save config information.
url = "https://apps.bea.gov/api/data/?UserID=" + api_key + "&method=GetData&datasetname=Regional&TableName=CAINC1&LineCode=3&Year=2017&GeoFips=COUNTY&ResultFormat=json"
units = "imperial"

# Build partial query URL
query_url = f"{url}"
```

```python
response = requests.get(query_url).json()
response
```

```python
df = pd.DataFrame(response['BEAAPI']['Results']['Data'])
```

```python
df
```

```python
df = df.join(df["GeoName"].str.split(",", 1, expand=True).rename(columns={0:'County', 1:'State Abbr'}))
df["County"] = df["County"].str.replace("County", "")
df.head()
```

```python
reduced_df = df.loc[:, ["State Abbr", "County", "DataValue"]]
reduced_df.rename(columns={"DataValue": "Per Capita Personal Income"}, inplace=True)
reduced_df = reduced_df[reduced_df["State Abbr"].astype(str) != "None"]

reduced_df.head()
```
