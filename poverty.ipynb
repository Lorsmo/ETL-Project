```python
import pandas as pd
from sqlalchemy import create_engine
from config import pwd
import psycopg2
```

### Store CSV into DataFrame

```python
csv_file = "Resources/allpovu.csv"
data_df = pd.read_csv(csv_file)
data_df.head()
```

### Create new data with select columns

```python
new_header = data_df.iloc[2] #grab the first row for the header
data_df.columns = new_header
data_df.head()
```

```python
poverty_data_df = data_df[4:]
poverty_data_df
```

```python
poverty_data_df = poverty_data_df.iloc[:, 0:8] # first seven columns of data frame with all rows
poverty_data_df
```

```python
poverty_data_df
```

```python
poverty_df = poverty_data_df[['State FIPS code', 'County FIPS code', 'Name','State Postal Code','Poverty Universe, All Ages']].copy()
poverty_df = poverty_df.loc[poverty_df['County FIPS code'] != '0']
poverty_df.head()
```

```python
def remove_county(county):
    return county.split(" ")[0]

poverty_df['Name'] = poverty_df['Name'].apply(remove_county)
poverty_df.head()
```

```python
poverty_df = poverty_df.reset_index()
poverty_df.head()
```

```python
poverty_df = poverty_df[['State FIPS code', 'County FIPS code', 'Name', 'State Postal Code', 'Poverty Universe, All Ages']]
poverty_df.head()
```

```python
poverty_df.columns = ['State FIPS code', 'County FIPS code', 'Name', 'State Postal Code', 'Poverty Universe, All Ages']
poverty_df = poverty_df.rename(columns={"State FIPS code": "state_fips_code",
                                                         "County FIPS code": "county_fips_code",
                                                         "Name": "name_county",
                                                         "State Postal Code": "state_abbr",
                                                          "Poverty Universe, All Ages": "poverty_population" })


poverty_df.head()
```

```python
poverty_df['poverty_population'] = poverty_df['poverty_population'].astype(str).str.replace(',', '')
poverty_df.count()
```

### Connect to local database

```python
rds_connection_string = f"postgres:{pwd}@localhost:5432/ETL"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
poverty_df.to_sql(name='poverty', con=engine, if_exists='append', index=False)
```
